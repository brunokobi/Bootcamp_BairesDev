## üìã Descri√ß√£o do Desafio  
Seguindo o exemplo do algoritmo de binariza√ß√£o apresentado em nossa √∫ltima aula, implemente um programa em **Python** üêç que realize as seguintes transforma√ß√µes em uma imagem colorida:  

1. **Converter a imagem para n√≠veis de cinza** (valores de 0 a 255).  
2. **Converter a imagem para bin√°ria** (preto e branco, com valores de 0 e 255).  

## üéØ Objetivo  
A partir da imagem de entrada (colorida), obtenha duas sa√≠das:  
- A imagem em n√≠veis de cinza.  
- A imagem binarizada.   

---

üöÄ **Boa sorte no desafio e m√£os √† obra!**  


```python
# Importa as bibliotecas necess√°rias
import cv2
import matplotlib.pyplot as plt
from skimage import data
import numpy as np
import time

# Carrega a imagem colorida utilizando o skimage
default_image = data.astronaut()  # Astronaut √© uma imagem colorida dispon√≠vel no skimage

# Medir o tempo de processamento para a imagem original colorida
start_time = time.time()
# Simula√ß√£o do processamento para a imagem original
color_process_time = time.time() - start_time

# Converte a imagem colorida para tons de cinza e mede o tempo
start_time = time.time()
gray_image = cv2.cvtColor(default_image, cv2.COLOR_RGB2GRAY)
gray_process_time = time.time() - start_time

# Converte a imagem em tons de cinza para bin√°ria e mede o tempo
start_time = time.time()
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
binary_process_time = time.time() - start_time

# Plota as imagens original, em tons de cinza e bin√°ria
plt.figure(figsize=(12, 8))

# Plota a imagem colorida
plt.subplot(2, 3, 1)
plt.imshow(default_image)  # A imagem j√° est√° em RGB
plt.axis('off')
plt.title("Imagem Colorida")

# Plota a imagem em tons de cinza
plt.subplot(2, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Imagem em Tons de Cinza")

# Plota a imagem bin√°ria
plt.subplot(2, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.title("Imagem Bin√°ria")

# Exibe o gr√°fico
plt.tight_layout()
plt.show()
```
